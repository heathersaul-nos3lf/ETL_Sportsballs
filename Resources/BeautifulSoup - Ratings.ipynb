```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.sportsmediawatch.com/2012/12/college-football-wrap-tv-ratings-for-almost-every-game-this-season/'
response = requests.get(url)
```

```python
soup = bs(response.text, 'lxml')
print(soup.prettify())
```

```python
rating_table = soup.find('table', id='smwtable3').find_all('tr')
```

```python
print(rating_table)
```

```python
for header in soup.find_all('th'):
    print(header.text)
```
